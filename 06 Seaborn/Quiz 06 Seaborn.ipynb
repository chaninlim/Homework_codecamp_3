```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#done
#1สร้าง DataFrame ที่เก็บยอดขายของบริษัท A,B,C,D ของปี 2014,2015,2016,2017 โดยชื่อบริษัทเป็น index 
#และปีเป็น column. สร้างคำสั่ง numpy สุ่มเลขแบบ float ให้เป็นรายได้ของบริษัทที่อยู่ในช่วง 50 – 500 (หน่วยล้านบาท) 

rows=['A','B','C','D']
#columns=['2014','2015','2016','2017']
columns=np.arange(2014,2018)

values=np.random.rand(4,4)
values2=values*100
values2
df=pd.DataFrame(values2,rows,columns)
df
```

```python
#2 done ใช้ข้อมูลจาก DataFrame ด้านบนสร้าง Bar plot ให้แกน x เป็น ชื่อบริษัท ยอดขายเฉลี่ยเป็นแกน Y โดยสร้างกราฟแบบแนวตั้งและแนวนอน

df['AvgSale']=(df[2014]+df[2015]+df[2016]+df[2017])/4
#workdf['AvgSale']=(df[2014]+100)

df['AvgSale']
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.barplot(x=df.index,y=df['AvgSale'])
plt.subplot(1,2,2)
sns.barplot(y=df.index,x=df['AvgSale'])
```

```python
#3ใdone ช้ข้อมูลจาก DataFrame ด้านบนสร้าง Scatter plot ให้แกน x เป็น ชื่อบริษัท ยอดขายปี 2017 เป็นแกน Y โดยสร้างกราฟแบบแนวตั้งและแนวนอน
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.scatterplot(x=df.index,y=df[2017],data=df)

plt.subplot(1,2,2)
sns.scatterplot(y=df.index,x=df[2017],data=df)
```

```python
#4.done สร้าง count plot นับข้อมูล Embarked โดยสร้างกราฟแบบแนวตั้งและแนวนอน
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
R41=sns.countplot(x=df['Embarked'])
plt.subplot(1,2,2)
R42=sns.countplot(y=df['Embarked'])
```

```python
#5 done สร้าง count plot นับข้อมูล Pclass โดยสร้างกราฟแบบแนวตั้งและแนวนอน
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.countplot(x=df['Pclass'])
plt.subplot(1,2,2)
sns.countplot(y=df['Pclass'])
```

```python
#done 6สร้าง box plot ให้แกน X เป็นเพศ และแกน Y เป็นราคา 
plt.figure(figsize=(4,6))
sns.boxplot(x='Sex',y='Fare',data=df)
#df.info()
```

```python
#done 7สร้าง box plot ให้แกน X เป็นPclass และแกน Y เป็นราคา 
plt.figure(figsize=(6,8))
sns.boxplot(x='Pclass',y='Fare',data=df)
```

```python
#8 done สร้าง barplot โดยให้ แกน X เป็นจุดหมายปลายทางและแกน Y เป็นราคา
sns.barplot(x='Embarked',y='Fare',data=df)
```

```python
#9 done สร้าง barplot โดยให้ แกน X เป็นPclass และแกน Y เป็นอายุ
sns.barplot(x='Pclass',y='Age',data=df)
```

```python
#10 done สร้าง Stripplot โดยให้แกน X เป็น Pclass และ แกน Y เป็นราคา
plt.figure(figsize=(4,6))
sns.stripplot(x='Pclass',y='Fare',data=df)
```

```python
#11 done สร้าง Stripplot โดยให้ แกน X เป็น Survived แกน Y เป็นอายุ และ hue เป็นเพศ
plt.figure(figsize=(6,8))
sns.stripplot(x='Survived',y='Age',data=df,hue='Sex')
```

```python
#12 done สร้าง Swarmplot โดยให้ แกน X เป็นเพศและแกน Y เป็นอายุ
plt.figure(figsize=(8,6))
sns.swarmplot(x='Sex',y='Age',data=df)
```

```python
#13 done สร้าง HeatMap โดย crosstab กำหนดให้ index เป็น Pclass และ column เป็นเพศ 
df13=pd.crosstab(df['Pclass'],df['Sex'])
df13
sns.heatmap(df13)
```

```python
#14 done สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา
df14=df.pivot_table(index='Pclass',columns='Sex',values='Fare')
df14
sns.heatmap(df14)
```

```python
#15 done สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา 
#และกำหนดเส้นสีขาวขนาด 2 เป็นเส้นแบ่ง
sns.heatmap(df14,linecolor='white',linewidth=2)
```

```python
#16 done สร้าง Heatmap โดย Correlation ของทั้ง DataFrame
df16_corr=df.corr()
df16_corr
sns.heatmap(df16_corr)
```

```python
#17 done สร้าง Clustermap ของทั้ง DataFrame
sns.clustermap(df16_corr)
```
