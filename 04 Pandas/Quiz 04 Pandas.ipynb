```python
#20. done แสดงเป็นกลุ่มช่วงอายุ 0-10, 20-30 …. (เพิ่มทีละ10) และหาว่าแต่ละกลุ่มมีอัตราการรอดชีวิตเท่าไหร่ 
df20=pd.read_csv('train.csv')
age_group=df20.head(891)['Age'].apply(lambda x:((x//10)*10))
age_group2=df20.head(891)['Age'].apply(lambda x:     ((x//10)*10)+9    )
age_group2
df20['Age_group']=pd.Series(age_group)
df20['Age_group2']=pd.Series(age_group2)
df20.sample(20)

df20['Age_group']=df20['Age_group'].astype(str)
df20['Age_group2']=df20['Age_group2'].astype(str)

df20['AgeGroup']=df20['Age_group']+'-'+df20['Age_group2']
df20

check0=df20.groupby(['AgeGroup'])
total=check0.describe()['Survived']['count']#count age group 

df2000=pd.DataFrame(total)
df2000.rename(columns={'count':'Total'},inplace=True)
df2000
df201=df20[df20['Survived']>0] # Survived only
check1=df201.groupby(['AgeGroup'])
CountSurvived=check1.describe()['Survived']['count']
df2000['Count Survived']=pd.Series(CountSurvived)
df2000['Survived Ratio(%)']=df2000['Count Survived']*100/df2000['Total']
df2000
```
