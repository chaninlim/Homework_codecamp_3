```python
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
วิเคราะห์ข้อมูลจากยอดขายของวิดีโอเกมส์ จากกว่า 16,500 เกมส์
Rank - Ranking of overall sales
Name - The games name
Platform - Platform of the games release (i.e. PC,PS4, etc.)
Year - Year of the game's release
Genre - Genre of the game
Publisher - Publisher of the game
NA_Sales - Sales in North America (in millions)
EU_Sales - Sales in Europe (in millions)
JP_Sales - Sales in Japan (in millions)
Other_Sales - Sales in the rest of the world (in millions)
Global_Sales - Total worldwide sales.
```

```python
#7. done หารายละเอียดเกมส์ GTA V ภาคต่างๆ ทั้งหมด 
#Grand Theft Auto V

df7 = df[df['Name'].str.contains('Grand Theft')]
df7.sort_values(by=['Name'])
#df7.groupby('Name').describe()
```

```python
#8.done หาจำนวนเกมส์ที่มีชื่อซ้ำกันทั้งหมด
df8=df['Name'].value_counts()
df81=df8[df8>1]
print('จำนวนเกมส์ที่มีชื่อซ้ำกันทั้งหมด=',len(df81))
(df81)
```

```python
#9.done หาจำนวนเกมส์ที่มีชื่อและ Platform ซ้ำกันทั้งหมด
df['Platform_Name']=df['Platform']+'-'+df['Name']
df9=df['Platform_Name'].value_counts()
df9.index
df92=df9[df9.values >1]
print('จำนวนเกมส์ที่มีชื่อและ Platform ซ้ำกันทั้งหมด=',df92.count())
df92
```

```python
#10. done จากข้อ 9 พิจารณาแถวที่ข้อมูลซ้ำแและลบแถวนั้น

type([df92.index]) # list
print(df.loc[df['Platform_Name'].isin([df92.index])])

#ใช้ code ด้านบน ไม่ได้ผลลัพ เลย ขอใช้ manual ตามด้านล่าง
list_to_find=['X360-Need for Speed: Most Wanted', 'Wii-Wii de Asobu: Metroid Prime',
       'PC-Need for Speed: Most Wanted', 'PS3-Madden NFL 13',
       'PS3-Sonic the Hedgehog']
print(df.loc[df['Platform_Name'].isin(list_to_find)])

#Need for Speed: Most Wanted ทั้ง 4 rows เป็นคนละปี จึงเก็บไว้
#Madden NFL 13    บน  PS3  เลือกลบ row ที่ค่า Global_Sales น้อย  rank 16130   id 16127
#ที่เหลือ ลบ row ที่มี ค่า Nan  rank 15002(id 14999)   4147 (id 4145)

#df.loc[df['Rank']==16130]
#df.loc[df['Rank']==15002]
#df.loc[df['Rank']==4147]

df.iloc[16127]
df.drop(16127,inplace=True) 
df.iloc[14999]
df.drop(14999,inplace=True) 
df.iloc[4145]
df.drop(4145,inplace=True) 
df
```

```python
#11. done หารายได้ทั้งหมดเกมส์ FIFA 15
#df11 = df[df['Name'].str.contains('FIFA')]
#df11['Name'].unique()
df11 = df[df['Name'].str.contains('FIFA 15')]
#df11.groupby('Name').describe()['Global_Sales']
df112=df11.pivot_table(index='Name',values='Global_Sales',aggfunc=np.sum)
df112
```

```python
#12. done หารายได้ทั้งหมดของเกมส์ GTA V ในญี่ปุ่น
df12 = df[df['Name'].str.contains('Grand Theft')]
df12
#df121=df12.pivot_table(index='Publisher',values='JP_Sales',aggfunc=np.sum)
df121=df12.pivot_table(index='Publisher',values='JP_Sales',aggfunc='sum')
df121
```

```python
#13. done สร้าง DF ที่ให้ index เป็นรายชื่อของเกมส์ที่มีซ้ำ และ Column เป็นจำนวนที่มีซ้ำ ไม่นับที่เป็น Unique
#df13=df.pivot_table(index='Name',values='Name',aggfunc=np.count)
df13=df.pivot_table(values='Rank', index='Name', aggfunc='count')
df13
df13.rename(columns={"Rank": "Count"},inplace=True)
df131=df13[df13['Count']>1]
df131.sort_values(by='Count',ascending=False)
```

```python
#14.done สร้าง DF ที่มี index เป็นชื่อบริษัท และเรียงลำดับตามรายได้รวมจากมากไปน้อย
df14=df.pivot_table(index='Publisher',values='Global_Sales',aggfunc=np.sum)
df14.sort_values(by='Global_Sales',ascending=False)
```

```python
#15.done จงสร้าง DF ที่บรรจุเกมส์ Series Call of duty ทั้งหมด
df15 = df[df['Name'].str.contains('Call of Duty')]
df15
```

```python
#16. done จงหาว่า Call of Duty ภาคใดใน PC มีรายสูงสุดในยุโรป 5 อันดับแรก
df16 = df15[df15['Platform'].str.contains('PC')]
df16.sort_values('EU_Sales')[::-1].head(5)
```

```python
#17. done จงหาว่า Platform ใด มียอดขายรวมสูงสุดในยุโรป
df17=df.pivot_table(index='Platform',values='EU_Sales',aggfunc=np.sum)
df17
df17.sort_values('EU_Sales')[::-1].head()
```

```python
#18. done จงหาว่าเกมส์ประเภทใดมียอดขายเฉลี่ยสูงสุดในภูมิภาคอื่นๆรอบโลก
df18=df.pivot_table(index='Genre',values='Other_Sales',aggfunc='mean')
df18.sort_values('Other_Sales')[::-1]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import plotly.express as px
```

```python
#19. done สร้าง Bar Plot โดยให้แกน X เป็น Platform และ Y เป็นยอดขายทั่วโลก

plt.figure(figsize=[15,5])
sns.barplot(x=df['Platform'],y=df['Global_Sales'])
```

```python
#20. done สร้าง Pie Chart หาส่วนแบ่งทางการตลาดของ 5 บริษัทแรกที่มีรายได้มากที่สุด
df20=df.pivot_table(index='Publisher',values='Global_Sales',aggfunc='sum')
df201=df20.sort_values('Global_Sales')[::-1].head(5)
plt.figure(figsize=(10,10))
plt.pie(df201['Global_Sales'],labels=df201.index, autopct='%1.1f%%')
```

```python
#21. doneสร้าง Count Plot นับข้อมูลประเภทของเกมส์
plt.figure(figsize=(20,8))
sns.countplot(x='Genre',data=df)
```

```python
#22. done สร้าง Bar plot Top 5 รายได้ทั่วโลกของ Call of Duty ภาคต่างๆใน Xbox
#df15['Platform'].unique()
df22=df15[df15['Platform'].isin(['X360', 'XOne'])]
df22
#df221=df22.pivot_table(index='Name',columns='drive-wheels',values='Global_Sales')
df221=df22.pivot_table(index='Name',values='Global_Sales',aggfunc='sum')
df222=df221.sort_values('Global_Sales')[::-1].head(5)
df222

fig=plt.figure(figsize=(10,8))
sns.barplot(x=df222.index,y=df222['Global_Sales'])
fig.autofmt_xdate()
```

```python
#23.done สร้าง Line Graph แสดงรายได้จาก North America จากปีแรกถึงปีสุดท้าย
df23=df[['Year','NA_Sales']]
df23
df231=df23.pivot_table(index='Year',values='NA_Sales',aggfunc='sum')
df231

fig=plt.figure(figsize=(10,8))
sns.lineplot(data=df231, x=df231.index, y='NA_Sales')
```

```python
#24. done สร้าง Stripplot โดยให้แกน X เป็น Genre และ แกน Y เป็นรายได้ทั่วโลก
df24=df.pivot_table(index='Genre',values='Global_Sales',aggfunc='sum')
df24
fig=plt.figure(figsize=(10,8))
sns.stripplot(x=df24.index,y='Global_Sales',data=df24)
fig.autofmt_xdate()
```

```python
#25. done สร้าง Distribution Plot ของปี
sns.distplot(df['Year'])
```
