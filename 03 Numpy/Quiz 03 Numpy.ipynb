```python
#18-3 จงเขียนโปรแกรมสุ่มเลขจำนวนเต็มในช่วง 10-100 ออกมาอยู่ใน matrix 5x5 
#และใช้ If-else เช็คทีละตัวว่าตัวใดบ้างที่มีค่ามากกว่า 50 และปริ้นออกมา
#
a=np.random.randint(10,101,25)
b=a.reshape(5,5)
print(b)
b[4][4]

lst=np.arange(0,5)
lst
print('result ตัวที่มากกว่า 50')
for x in lst:
    for y in lst:
        if b[x][y] >50:
            print(b[x][y])
        y=y+1
    x=x+1
```
