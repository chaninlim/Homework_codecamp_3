```python
#1 done สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน 
import numpy as np
import matplotlib.pyplot as plt
# to view in jp notebook
%matplotlib inline


x=np.arange(1,20,2)
y=x^2

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('X ^ 2')
plt.title('X ^ 2 Chart')
plt.show()
```

```python
#2 สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีชมพู 
x=np.arange(0,90)
y=np.sin(x)
plt.plot(x,y,'m')
plt.xlabel('X axis')
plt.ylabel('Y axis - Sin')
plt.title('Quiz 2 - Sin of X')
plt.show()
```

```python
#3 สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก
y=np.arange(1,10)
x=y^2
plt.plot(x,y,c='#66FF66')
plt.xlabel('Y ^ 2')
plt.ylabel('Y axis')
plt.title("Quiz 3")
```

```python
#4 สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - - 
x=np.arange(1,10)
y=np.cos(x)
plt.plot(x,y,'r',linestyle='--',)
plt.xlabel('X axis')
plt.ylabel('Cos')
plt.title('Quiz 4 - Cos')
plt.show()
```

```python
#5. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน 
#ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker
x=np.arange(1,20)
y1=x^2
y2=x^3

plt.figure(figsize=[20,5])

plt.subplot(1,2,1)
plt.plot(x,y1,'r',marker='o')
plt.xlabel('X1')
plt.ylabel('X ^ 2')
plt.title('Quiz5-1')

plt.subplot(1,2,2)
plt.plot(x,y2,'b',marker='^')
plt.xlabel('X2')
plt.ylabel('X ^ 3')
plt.title('Quiz5-2')

plt.show()
```

```python
#6. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน 
#พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน 
#ใช้สีแดงและน้ำเงินตามลำดับ 
#โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker 
#และใส่ legend 
#และเพิ่มเส้นตารางข้างหลัง
x=np.arange(1,50)
y=x^2
x=y^2

plt.figure(figsize=[20,8])
plt.subplot(1,2,1)
plt.plot(x,y,'r', marker='o')
plt.title('Q6-1')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.legend('legend1')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,y,'b', marker='^')
plt.title('Q6-2')
plt.xlabel('X2')
plt.ylabel('Y2')
plt.legend('legend2')
plt.grid()
```

```python
#7. สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม 
y=np.arange(1,10)
x=y^2 + 4*y
plt.plot(x,y,c='#CCCC66')
plt.xlim(15,40)
plt.ylim(3,8)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph 7 : function X = Y^2 + 4*Y')
plt.show
```

```python
#8. สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend 
y=np.arange(1,30)
x=y^3

fig=plt.figure(figsize=[10,5])
ax1=fig.add_axes([.1,.1,.9,.9])
ax1.plot(x,y,label='legend 1')
ax1.set_xlabel('Y ^3')
ax1.set_ylabel('Y1')
ax1.set_title('Quiz 8-1')
ax1.legend(loc=2)

x=np.arange(1,30)
y=x^2+2

ax2=fig.add_axes([.6,.2,.3,.2])
ax2.plot(x,y, label='legend 2')

ax2.set_xlabel('X2')
ax2.set_ylabel('X^2 +2')
ax2.set_title('Quiz 8-2')
ax2.legend(loc=0)
```
